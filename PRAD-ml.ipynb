```python
# Función para filtrar genes con baja expresión y normalizar los datos
def filter_and_normalize(data, expression_threshold=1, sample_threshold=0.1):
    metastatic_column = data['metastatic']
    data_without_metastatic = data.drop(columns=['metastatic', 'barcode'])
    # Separar datos de metástasis y no metástasis
    metastatic_data = data_without_metastatic[metastatic_column == 1]
    non_metastatic_data = data_without_metastatic[metastatic_column == 0]

    # Filtrar genes con baja expresión en datos de metástasis
    num_samples_metastatic = metastatic_data.shape[0]
    metastatic_filtered = metastatic_data.loc[:, (metastatic_data > expression_threshold).sum(axis=0) > sample_threshold * num_samples_metastatic]

    # Filtrar genes con baja expresión en datos de no metástasis
    num_samples_non_metastatic = non_metastatic_data.shape[0]
    non_metastatic_filtered = non_metastatic_data.loc[:, (non_metastatic_data > expression_threshold).sum(axis=0) > sample_threshold * num_samples_non_metastatic]

    # Obtener los nombres de los genes que cumplen con los criterios en cualquiera de los dos grupos
    filtered_genes = list(set(metastatic_filtered.columns) | set(non_metastatic_filtered.columns))

    # Filtrar el dataset original para mantener solo los genes importantes
    data_filtered = data_without_metastatic[filtered_genes]

    # Normalización logarítmica
    data_normalized = data_filtered.applymap(lambda x: np.log2(x + 1))

    data_normalized.insert(0, 'metastatic', metastatic_column)
    return data_normalized

# Normalizar y filtrar los datos combinados
normalized_genes = filter_and_normalize(merged_genes)
```

```python
# Crear el generador
def build_generator(input_dim, output_dim):
    model = Sequential()
    model.add(Input(shape=(input_dim,)))
    model.add(Dense(128))
    model.add(LeakyReLU(alpha=0.02))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(256))
    model.add(LeakyReLU(alpha=0.02))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(512))
    model.add(LeakyReLU(alpha=0.02))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(output_dim, activation='tanh'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Crear el discriminador
def build_discriminator(input_dim):
    model = Sequential()
    model.add(Input(shape=(input_dim,)))
    model.add(Dense(512))
    model.add(LeakyReLU(alpha=0.02))
    model.add(Dense(256))
    model.add(LeakyReLU(alpha=0.02))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Parámetros
data_dim = X_train_minority_np.shape[1]
```

```python
# Instanciar generador y discriminador
generator = build_generator(input_dim, data_dim)
discriminator = build_discriminator(data_dim)
```

```python
def build_GAN(generator, discriminator):
    model = Sequential()
    model.add(generator)
    discriminator.trainable = False
    model.add(discriminator)
    model.compile(optimizer='adam', loss='binary_crossentropy')
    return model

gan = build_GAN(generator, discriminator)
```

```python
batch_size = 12
iterations = 5000
real = np.ones(batch_size)*0.9
fake = np.zeros(batch_size)*0.1
sample_interval = 500
```

```python
tf.keras.backend.clear_session()
```

```python
end_loop = 0
for i in range(1,iterations+1):
    print("Epoch " + str(i))

    noise = np.random.normal(0,1,[batch_size,input_dim])
    batch_fake = generator.predict(noise)

    idx = np.random.randint(low=0, high=X_train_minority_np.shape[0],size=batch_size)
    batch_real = X_train_minority_np[idx]

    discriminator.trainable = True

    d_loss_real = discriminator.train_on_batch(batch_real,real)
    d_loss_fake = discriminator.train_on_batch(batch_fake,fake)
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    discriminator.trainable = False

    noise = np.random.normal(0, 1, (batch_size, input_dim))
    g_loss = gan.train_on_batch(noise, np.ones(batch_size))

    if (i) % sample_interval == 0:
        print(f"{i}/{i-1} [D loss: {d_loss[0]} | D accuracy: {100*d_loss[1]}] [G loss: {g_loss}]")

    if abs(50-100*d_loss[1])<1:
      end_loop = end_loop + 1

    del noise, batch_fake, batch_real, d_loss_real, d_loss_fake, d_loss, g_loss, idx

    if end_loop==50:
      break
```

# 2.1 Regresión Logística

```python
# Definir una función para entrenar y evaluar modelos
def train_lr_model(X_train, y_train, X_val, y_val, X_test, y_test, method_name, save_directory):
    X_train_val = np.concatenate((X_train, X_val), axis=0)
    y_train_val = np.concatenate((y_train, y_val), axis=0)

    # Definir el modelo
    cv = KFold(n_splits=5, random_state=my_seed, shuffle=True)
    lr_clf = LogisticRegression(penalty='l2', random_state=my_seed)

    # Definir los parámetros de hiperparámetros para la búsqueda
    param_grid = {
        'max_iter': [100, 500, 1000],
        'C': [0.01, 0.1, 1, 10, 100]
    }

    # Usar f1-score para la clase 1 como métrica
    scorer = make_scorer(f1_score, pos_label=1)

    # Configurar GridSearchCV
    grid_search = GridSearchCV(lr_clf, param_grid, cv=cv, scoring=scorer)

    # Medir el tiempo de entrenamiento
    start_train_time = time.time()
    grid_search.fit(X_train_val, y_train_val)
    end_train_time = time.time()
    training_time = end_train_time - start_train_time

    # Obtener el mejor modelo encontrado
    best_model = grid_search.best_estimator_

    # Medir el tiempo de predicción
    start_predict_time = time.time()
    y_test_pred = best_model.predict(X_test)
    end_predict_time = time.time()
    prediction_time = end_predict_time - start_predict_time

    print(f"Regresión Logística {method_name} - Prueba:")
    print(classification_report(y_test, y_test_pred, target_names=class_names))

    # Obtener las probabilidades de predicción para el conjunto de prueba
    y_test_proba = best_model.predict_proba(X_test)[:, 1]
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_proba)
    roc_auc_test = roc_auc_score(y_test, y_test_proba)

    # Visualizar la curva ROC para el conjunto de prueba
    plot_roc_curve(fpr_test, tpr_test, roc_auc_test, 'Prueba')

    # Visualizar la matriz de confusión en prueba
    plot_confusion_matrix(y_test, y_test_pred, class_names)

    # Número de predicciones correctas para la clase 1 en el conjunto de prueba
    correct_class_1_predictions = ((y_test == 1) & (y_test_pred == 1)).sum()
    total_class_1 = (y_test == 1).sum()
    print(f'Número de predicciones correctas para la clase 1 en prueba: {correct_class_1_predictions} de {total_class_1}')

    # Guardar el mejor modelo
    model_path = f'{save_directory}-lr_clf-{method_name}'
    joblib.dump(best_model, model_path)

    # Historial de grid_search
    results_df = pd.DataFrame(grid_search.cv_results_)
    selected_columns = [
        'param_max_iter', 'param_C', 'mean_test_score', 'std_test_score', 'rank_test_score'
    ]
    results_df = results_df[selected_columns].sort_values(by='rank_test_score')
    print(f'Best Parameters: {grid_search.best_params_}')
    print(results_df)

    print(f"Tiempo de entrenamiento: {training_time:.2f} segundos")
    print(f"Tiempo de predicción: {prediction_time:.2f} segundos")
```

```python
def train_nb_model(X_train, y_train, X_val, y_val, X_test, y_test, method_name, save_directory):
    X_train_val = np.concatenate((X_train, X_val), axis=0)
    y_train_val = np.concatenate((y_train, y_val), axis=0)

    # Definir el modelo
    nb_clf = GaussianNB()

    # Medir el tiempo de entrenamiento
    start_train_time = time.time()
    nb_clf.fit(X_train_val, y_train_val)
    end_train_time = time.time()
    training_time = end_train_time - start_train_time

    # Predicción y evaluación en el conjunto de validación
    y_val_pred = nb_clf.predict(X_val)
    print(f"Gaussian Naive Bayes {method_name} - Validación:")
    print(classification_report(y_val, y_val_pred, target_names=class_names))

    # Obtener las probabilidades de predicción para el conjunto de validación
    y_val_proba = nb_clf.predict_proba(X_val)[:, 1]
    fpr_val, tpr_val, _ = roc_curve(y_val, y_val_proba)
    roc_auc_val = roc_auc_score(y_val, y_val_proba)

    # Visualizar la curva ROC para el conjunto de validación
    plot_roc_curve(fpr_val, tpr_val, roc_auc_val, 'Validación')

    # Visualizar la matriz de confusión en validación
    plot_confusion_matrix(y_val, y_val_pred, class_names)

    # Predicción y evaluación en el conjunto de prueba
    start_predict_time = time.time()
    y_test_pred = nb_clf.predict(X_test)
    end_predict_time = time.time()
    prediction_time = end_predict_time - start_predict_time
    print(f"Gaussian Naive Bayes {method_name} - Prueba:")
    print(classification_report(y_test, y_test_pred, target_names=class_names))

    # Obtener las probabilidades de predicción para el conjunto de prueba
    y_test_proba = nb_clf.predict_proba(X_test)[:, 1]
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_proba)
    roc_auc_test = roc_auc_score(y_test, y_test_proba)

    # Visualizar la curva ROC para el conjunto de prueba
    plot_roc_curve(fpr_test, tpr_test, roc_auc_test, 'Prueba')

    # Visualizar la matriz de confusión en prueba
    plot_confusion_matrix(y_test, y_test_pred, class_names)

    # Número de predicciones correctas para la clase 1 en el conjunto de prueba
    correct_class_1_predictions = ((y_test == 1) & (y_test_pred == 1)).sum()
    total_class_1 = (y_test == 1).sum()
    print(f'Número de predicciones correctas para la clase 1 en prueba: {correct_class_1_predictions} de {total_class_1}')

    # Guardar el modelo entrenado
    model_path = f'{save_directory}-nb_clf-{method_name}'
    joblib.dump(nb_clf, model_path)


    print(f"Tiempo de entrenamiento: {training_time:.2f} segundos")
    print(f"Tiempo de predicción: {prediction_time:.2f} segundos")
```

```python
# Visualizar la pérdida del modelo XGBoost
def plot_loss_acc_xgb(evals_result):
    epochs = len(evals_result['validation']['logloss'])
    x_axis = range(0, epochs)
    plt.figure(figsize=(12, 6))
    plt.plot(x_axis, evals_result['train']['logloss'], label='Train')
    plt.plot(x_axis, evals_result['validation']['logloss'], label='Validation')
    plt.title('Model logloss')
    plt.ylabel('Logloss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.show()
```

```python
def train_xgb_model(X_train, y_train, X_val, y_val, X_test, y_test, method_name, save_directory):
    def objective(trial):
        # Definir los parámetros de hiperparámetros para la búsqueda
        params = {
            'objective': 'binary:logistic',
            'max_depth': trial.suggest_int('max_depth', 3, 5),
            'eta': trial.suggest_float('eta', 0.01, 0.3, log=True),
            'eval_metric': 'logloss',
            'seed': my_seed
        }

        def f1_eval(preds, dtrain):
            labels = dtrain.get_label()
            preds = (preds > 0.5).astype(int)
            f1 = f1_score(labels, preds, pos_label=1)
            return 'f1', f1

        dtrain = xgb.DMatrix(X_train, label=y_train)
        dval = xgb.DMatrix(X_val, label=y_val)

        # Entrenamiento
        evals_result = {}
        bst = xgb.train(params, dtrain, num_boost_round=100, evals=[(dtrain, 'train'), (dval, 'validation')],
                        early_stopping_rounds=10, evals_result=evals_result, verbose_eval=False, custom_metric=f1_eval)

        # Predicción y evaluación en el conjunto de validación
        y_val_pred = (bst.predict(dval) > 0.5).astype(int)
        f1 = f1_score(y_val, y_val_pred, pos_label=1)
        return f1

    def f1_eval(preds, dtrain):
        labels = dtrain.get_label()
        preds = (preds > 0.5).astype(int)
        f1 = f1_score(labels, preds, pos_label=1)
        return 'f1', f1

    # Optimización de hiperparámetros con Optuna
    start_optuna_time = time.time()
    study = optuna.create_study(direction='maximize')
    study.optimize(objective, n_trials=5)
    end_optuna_time = time.time()
    optuna_search_time = end_optuna_time - start_optuna_time
    best_params = study.best_params
    best_params['eval_metric'] = 'logloss'

    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)
    dtest = xgb.DMatrix(X_test, label=y_test)
    evals_result = {}

    # Entrenamiento
    start_train_time = time.time()
    bst = xgb.train(best_params, dtrain, num_boost_round=100, evals=[(dtrain, 'train'), (dval, 'validation')],
                    early_stopping_rounds=10, evals_result=evals_result, custom_metric=f1_eval)
    end_train_time = time.time()
    training_time = end_train_time - start_train_time

    # Predicción
    start_predict_time = time.time()
    y_test_pred = (bst.predict(dtest) > 0.5).astype(int)
    end_predict_time = time.time()
    prediction_time = end_predict_time - start_predict_time

    # Predicción y evaluación en el conjunto de validación
    y_val_pred = (bst.predict(dval) > 0.5).astype(int)
    print(f"XGBoost {method_name} - Validación:")
    print(classification_report(y_val, y_val_pred, target_names=class_names))
    print(f'F1 Score (Clase 1): {f1_score(y_val, y_val_pred, pos_label=1):.4f}')

    # Obtener las probabilidades de predicción para el conjunto de validación
    y_val_proba = bst.predict(dval)
    fpr_val, tpr_val, _ = roc_curve(y_val, y_val_proba)
    roc_auc_val = roc_auc_score(y_val, y_val_proba)

    # Visualizar la curva ROC para el conjunto de validación
    plot_roc_curve(fpr_val, tpr_val, roc_auc_val, 'Validación')

    # Visualizar la matriz de confusión en validación
    plot_confusion_matrix(y_val, y_val_pred, class_names)

    # Predicción y evaluación en el conjunto de prueba
    y_test_pred = (bst.predict(dtest) > 0.5).astype(int)
    print(f"XGBoost {method_name} - Prueba:")
    print(classification_report(y_test, y_test_pred, target_names=class_names))
    print(f'F1 Score (Clase 1): {f1_score(y_test, y_test_pred, pos_label=1):.4f}')

    # Obtener las probabilidades de predicción para el conjunto de prueba
    y_test_proba = bst.predict(dtest)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_proba)
    roc_auc_test = roc_auc_score(y_test, y_test_proba)

    # Visualizar la curva ROC para el conjunto de prueba
    plot_roc_curve(fpr_test, tpr_test, roc_auc_test, 'Prueba')

    # Visualizar la matriz de confusión en prueba
    plot_confusion_matrix(y_test, y_test_pred, class_names)

    # Número de predicciones correctas para la clase 1 en el conjunto de prueba
    correct_class_1_predictions = ((y_test == 1) & (y_test_pred == 1)).sum()
    total_class_1 = (y_test == 1).sum()
    print(f'Número de predicciones correctas para la clase 1 en prueba: {correct_class_1_predictions} de {total_class_1}')

    # Guardar el modelo
    model_path_json = f'{save_directory}-xgb_clf-{method_name}.json'
    bst.save_model(model_path_json)

    # Historial de resultados
    plot_loss_acc_xgb(evals_result)
    results_df = pd.DataFrame({
        'epoch': list(range(len(evals_result['train']['logloss']))),
        'train_logloss': evals_result['train']['logloss'],
        'validation_logloss': evals_result['validation']['logloss']
    })
    print(f'Best Parameters: {best_params}')
    print(results_df)

    # Historial de parámetros
    trials_df = study.trials_dataframe(attrs=('number', 'value', 'params', 'state'))
    print(trials_df)

    print(f"Tiempo de búsqueda de parámetros Optuna: {optuna_search_time:.2f} segundos")
    print(f"Tiempo de entrenamiento: {training_time:.2f} segundos")
    print(f"Tiempo de predicción: {prediction_time:.2f} segundos")
```

```python
from optuna.integration import KerasPruningCallback

# Función de objetivo para Optuna
def objective(trial, X_train, y_train, X_val, y_val):
    # Hiperparámetros
    dropout_rate = trial.suggest_float('dropout_rate', 0.1, 0.5)
    layer1 = trial.suggest_int('layer1', 512, 1024)
    layer2 = trial.suggest_int('layer2', 256, layer1)
    layer3 = trial.suggest_int('layer3', 128, layer2)
    layer4 = trial.suggest_int('layer4', 64, layer3)
    layer5 = trial.suggest_int('layer5', 32, layer4)

    model_DNN_st = Sequential([
        Dense(layer1, activation='relu', input_shape=(X_train.shape[1],)),
        Dropout(dropout_rate),
        Dense(layer2, activation='relu'),
        Dropout(dropout_rate),
        Dense(layer3, activation='relu'),
        Dropout(dropout_rate),
        Dense(layer4, activation='relu'),
        Dropout(dropout_rate),
        Dense(layer5, activation='relu'),
        Dropout(dropout_rate),
        Dense(1, activation='sigmoid')
    ])

    model_DNN_st.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    pruning_callback = KerasPruningCallback(trial, 'val_loss')

    # Entrenamiento
    history = model_DNN_st.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val),
                               callbacks=[early_stop, pruning_callback], verbose=0)

    # Evaluar el modelo en el conjunto de validación
    y_val_pred_proba = model_DNN_st.predict(X_val)
    y_val_pred = (y_val_pred_proba > 0.5).astype(int)

    # Calcular el f1_score para la clase 1
    f1 = f1_score(y_val, y_val_pred, pos_label=1)

    tf.keras.backend.clear_session()

    return f1

# Función para entrenar el modelo DNN con los mejores hiperparámetros
def train_dnn_model(X_train, y_train, X_val, y_val, X_test, y_test, method_name, save_directory):
    # Crear estudio de Optuna
    start_optuna_time = time.time()
    study = optuna.create_study(direction='maximize')
    study.optimize(lambda trial: objective(trial, X_train, y_train, X_val, y_val), n_trials=15)
    end_optuna_time = time.time()
    optuna_search_time = end_optuna_time - start_optuna_time

    print(f"Best trial: {study.best_trial.value}")
    print("Best hyperparameters: ", study.best_trial.params)

    # Entrenar el modelo con los mejores hiperparámetros encontrados
    best_params = study.best_trial.params
    dropout_rate = best_params['dropout_rate']
    layer1 = best_params['layer1']
    layer2 = best_params['layer2']
    layer3 = best_params['layer3']
    layer4 = best_params['layer4']
    layer5 = best_params['layer5']

    model_DNN_st = Sequential([
        Dense(layer1, activation='relu', input_shape=(X_train.shape[1],)),
        Dropout(dropout_rate),
        Dense(layer2, activation='relu'),
        Dropout(dropout_rate),
        Dense(layer3, activation='relu'),
        Dropout(dropout_rate),
        Dense(layer4, activation='relu'),
        Dropout(dropout_rate),
        Dense(layer5, activation='relu'),
        Dropout(dropout_rate),
        Dense(1, activation='sigmoid')
    ])

    model_DNN_st.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    # Medir el tiempo de entrenamiento
    start_train_time = time.time()
    history = model_DNN_st.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val),
                               callbacks=[early_stop])
    end_train_time = time.time()
    training_time = end_train_time - start_train_time

    # Guardar el modelo entrenado
    model_path = f'{save_directory}-dnn_clf-{method_name}.keras'
    model_DNN_st.save(model_path)

    # Medir el tiempo de predicción
    start_predict_time = time.time()
    y_test_pred_proba = model_DNN_st.predict(X_test)
    y_test_pred = (y_test_pred_proba > 0.5).astype(int)
    end_predict_time = time.time()
    prediction_time = end_predict_time - start_predict_time

    # Evaluar el modelo en el conjunto de prueba
    loss, accuracy = model_DNN_st.evaluate(X_test, y_test)
    print(f'Test Accuracy: {accuracy:.4f}')

    print(f"DNN {method_name} - Prueba:")
    print(classification_report(y_test, y_test_pred))

    # Visualizar la curva ROC
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_pred_proba)
    roc_auc_test = roc_auc_score(y_test, y_test_pred_proba)
    print(f'AUC-ROC: {roc_auc_test:.4f}')
    plot_roc_curve(fpr_test, tpr_test, roc_auc_test, 'Prueba')

    # Visualizar la matriz de confusión
    plot_confusion_matrix(y_test, y_test_pred, class_names)

    # Número de predicciones correctas para la clase 1 en prueba
    correct_class_1_predictions = ((y_test == 1) & (y_test_pred == 1)).sum()
    total_class_1 = (y_test == 1).sum()
    print(f'Número de predicciones correctas para la clase 1 en prueba: {correct_class_1_predictions} de {total_class_1}')

    # Historial de entrenamiento
    history_df = pd.DataFrame(history.history)
    print(history_df)

    # Visualizar la pérdida y la precisión a lo largo del tiempo
    plot_loss_acc(history)

    # Historial de parámetros
    trials_df = study.trials_dataframe(attrs=('number', 'value', 'params', 'state'))
    print(trials_df)

    print(f"Tiempo de búsqueda de parámetros Optuna: {optuna_search_time:.2f} segundos")
    print(f"Tiempo de entrenamiento: {training_time:.2f} segundos")
    print(f"Tiempo de predicción: {prediction_time:.2f} segundos")
```

```python
train_data_st = pd.DataFrame(X_train_st)
train_data_st['metastatic'] = y_train_st

train_data_ada = pd.DataFrame(X_train_ada)
train_data_ada['metastatic'] = y_train_ada

train_data_gan = pd.DataFrame(X_train_gan)
train_data_gan['metastatic'] = y_train_gan

val_data = pd.DataFrame(X_val)
val_data['metastatic'] = y_val

test_data = pd.DataFrame(X_test)
test_data['metastatic'] = y_test

label = 'metastatic'
```

```python
def train_tabular_predictor_model(train_data, val_data, test_data, method_name, save_directory):
    model_path = f'{save_directory}-tabular_predictor-{method_name}'
    # Crear el predictor
    start_train_time = time.time()
    predictor = TabularPredictor(label=label, path=model_path).fit(train_data, tuning_data=val_data, time_limit=3600)
    end_train_time = time.time()
    training_time = end_train_time - start_train_time

    # Guardar el modelo
    predictor.save(model_path)

    # Evaluar el modelo en el conjunto de prueba
    start_predict_time = time.time()
    y_pred = predictor.predict(test_data)
    end_predict_time = time.time()
    prediction_time = end_predict_time - start_predict_time
    accuracy = np.mean(y_pred == y_test)

    # Informe de clasificación
    print(f"TabularPredictor {method_name} - Prueba:")
    print(classification_report(y_test, y_pred))

    # Calcular la AUC-ROC
    roc_auc_test = roc_auc_score(y_test, y_pred)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_pred)
    plot_roc_curve(fpr_test, tpr_test, roc_auc_test, 'Prueba')

    plot_confusion_matrix(y_test, y_pred, class_names)

    # Número de predicciones correctas para la clase 1 en prueba
    correct_class_1_predictions = ((y_test == 1) & (y_pred == 1)).sum()
    total_class_1 = (y_test == 1).sum()
    print(f'Número de predicciones correctas para la clase 1 en prueba: {correct_class_1_predictions} de {total_class_1}')

    print(f"Tiempo de entrenamiento: {training_time:.2f} segundos")
    print(f"Tiempo de predicción: {prediction_time:.2f} segundos")
```

```python
train_tabular_predictor_model(train_data_st, val_data, test_data, "ST", save_directory)
```
